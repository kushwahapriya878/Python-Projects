```python
#converting data datatype of rate column
def handleRate(x):
  x=str(x).split('/')
  x=x[0]
  return float(x)

df['rate']=df['rate'].apply(handleRate)
```



*   There are no missing value present in the data set



#Exploring listed in type

```python
plt.figure(figsize=(8,4))
sns.countplot(data=df, x='listed_in(type)')
plt.xlabel('Type of Restaurant')
plt.show()
```

**Insight**


*   Mejority of restaurant fall into Dining category



# Total votes by restaurant type

```python
restaurant_votes_df=df.groupby(by='listed_in(type)')['votes'].sum()
restaurant_votes_df=pd.DataFrame({'votes':restaurant_votes_df})
restaurant_votes_df
```

```python
plt.figure(figsize=(8,4))
plt.plot(restaurant_votes_df, c='green', marker='o')
plt.xlabel('Type of Restaurant', c='black', size=20)
plt.ylabel('Votes', c='black', size=20)
plt.show()
```

**Insight**
*  As per above analysis Dining type restaurant is more prefered by users.

# Checking online and offline order

```python
plt.figure(figsize=(5,3))
sns.countplot(data=df, x='online_order')
plt.xlabel('Online and Ofline order count', c='black', size=15)
plt.show()
```

 **Insight**

*  As per above analysis most restaurant are not accepting online order.

# Checking distribution of rating

```python
plt.figure(figsize=(8,4))
sns.histplot(data=df, x='rate', bins=10)
plt.title('Rating Distribution')
plt.show()
```

**Insight**
*  Mejority of restaurant received rating range between 3.5 to 4

# Checking approx_cost(for two people)

```python
df['approx_cost(for two people)'].value_counts()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(data=df, x='approx_cost(for two people)')
plt.title('Cost per pair')
plt.show()
```

**Insight**

* Majority of pairs like restaurant approx cost 300 follwed by 200,150, 400.

# Checking whether online order receive higher rating or offline order?

```python
plt.figure(figsize=(8,4))
sns.boxplot(data=df, x='online_order', y='rate')
plt.show()
```

**Insight**
*  As per the above analysis online orders are recieveing higher rating than offline order

# Checking correlation between features

```python
df.drop(columns=['name', 'online_order', 'book_table', 'listed_in(type)']).corr()
```

```python
pivot_table = df.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```

**Insight**
* As per above analysis Dining restaurant first accept offline order, Cafes are first accept online oredr
